```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_xml('datafile.xml')
```

```python
df.columns
```

```python
df[df['GROUP'] == 'Synthetic Fibres']
```

```python
df['GROUP'].value_counts().plot(kind = 'barh')
```

```python
latest_data = df[['PRODUCT','_2021-22_QTY_','_2021-22__VAL_' ]]
latest_data
```

```python
fm = latest_data.sort_values("_2021-22_QTY_", ascending=False).head(10)
fm
```

```python
fm.dtypes
```

```python
fm.plot(kind = 'bar', x = 'PRODUCT',y = ['_2021-22_QTY_', '_2021-22__VAL_'])
# set the title of the legend
title = 'Sales Data'

# get the artist objects
bars = plt.gca().get_children()

# create the legend
plt.legend([bars[0], bars[1]], ['_2021-22_QTY_', '_2021-22__VAL_'], title=title)

# display the plot
plt.show()
```
